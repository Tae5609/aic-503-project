```python
# Parameters
SAMPLING_PERCENTAGE = 0.06
DATASET_MAP = {
    "train": { "label": "train_label.csv", "dir": "./images-train" },
    "test": { "label": "test_label.csv", "dir": "./images-test" },
    "val": { "label": "val_label.csv", "dir": "./images-val" },
}
OUTPUT_DIRECTORY = "./samples"
HEADER_HEIGHT = 60
RANDOM_SEED = 42
```

## Randomly Select

```python
import os
os.makedirs(OUTPUT_DIRECTORY, exist_ok=True)

# Read image files
n = 0
images = []

for dataset in DATASET_MAP:
    files = os.listdir(DATASET_MAP[dataset]['dir'])
    n += len(files)

    images += [(dataset, f"{DATASET_MAP[dataset]['dir']}/{filename}") for filename in files]

print(f"Total (N): {n}")
print(f"First 10 images: {images[:10]}")
```

```python
# Calculate number of samples
n_samples = int(0.06 * n)
print(f"Number of samples (Ns): {n_samples}")
```

```python
import random

# Randomly pick it (stratified)
indices = list(range(0, n))

sample = []
for i in range(n_samples):
    start = int(i * len(indices) / n_samples)
    end = int((i + 1) * len(indices) / n_samples)
    bucket = indices[start:end]
    if RANDOM_SEED: random.seed(RANDOM_SEED)
    sample.append(random.choice(bucket))

print(f"Sample length: {len(sample)}")
print(f"Sample: {sample}")
```

```python
# Map indices back to images
sample_images = []
for idx in sample:
    sample_images.append(images[idx])

print(f"First 10 sample image: {sample_images[:10]}")
```

## Generate Sample Images with label

```python
import csv

# Load the label
label_maps = {}

for dataset, config in DATASET_MAP.items():
    label_map = {}
    with open(config["label"], newline="") as f:
        reader = csv.reader(f)
        for row in reader:
            filename, label = row[0], row[1]
            label_map[filename] = label
    label_maps[dataset] = label_map
```

```python
from PIL import Image, ImageDraw, ImageFont

# Generate the sample images
font = ImageFont.load_default()

for dataset, img_path in sample_images:
    filename = os.path.basename(img_path)

    # Load image
    img = Image.open(img_path).convert("RGB")
    width, height = img.size

    # Create new image with header
    new_img = Image.new("RGB", (width, height + HEADER_HEIGHT), (30, 30, 30))
    new_img.paste(img, (0, HEADER_HEIGHT))

    # Draw header text
    draw = ImageDraw.Draw(new_img)
    label = label_maps[dataset].get(filename, "UNKNOWN")
    text = f"[{dataset.upper()}] {label}"
    draw.text((10, 30), text, font=font, fill=(255, 255, 255))

    # Save
    output_path = os.path.join(OUTPUT_DIRECTORY, f"{dataset}_{filename}")
    new_img.save(output_path)
```
